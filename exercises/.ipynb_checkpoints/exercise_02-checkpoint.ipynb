**4. Construct three files from ex1.dat (using already learned UNIX techniques). Each file should contain one column of numbers from ex1.dat. Now make a Python program that sums numbers from a input file and displays the sum of the numbers. Use the 3 files you constructed as input filers and see the sums. The sums are approx. Col 1; -904.4143, Col 2; 482.8410, Col 3; 292.05150 for the three columns.**

```julia
open("/Users/laurasansc/github/learning_julia/files/ex13.dat") do file
    n = 0
    for line in eachline(file)
        n += parse(Float64, line)
    end
    println(n)
end
```

**5. Based on the previous program, now make a new one that calculates the mean value of the columns. Also create a file with the numbers 9-1 and feed that to the program. What is the result from your program and what should it be, based on calculations in your head?**

```julia
open("/Users/laurasansc/github/learning_julia/files/ex13.dat") do file
    n = 0
    n_line = 0
    for line in eachline(file)
        n += parse(Float64, line)
        n_line += 1
    end
    println("Sum:", n)
    println("Number of lines", n_line)
    println("Average:", n / n_line )
end
```
