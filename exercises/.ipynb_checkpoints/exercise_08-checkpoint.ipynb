**1. Make a function that take a DNA sequence (string) as parameter and return the complement strand (reverse complement). Use it to improve 7.3 which works on dna7.fsa.**

```julia
#infile = ARGS[1]
#println("Which is your file?")
#infile = readline()

# Since there is not a function " make trans " in julia
# make a function that does it

function transcribe(sequence::String)
    complement = ""
    sequence = uppercase(sequence)
    for letter in sequence
        if string(letter) == "A"
            complement = string(complement,"T")
        elseif string(letter) == "T"
            complement = string(complement,"A")
        elseif string(letter) == "C"
            complement = string(complement,"G")
        elseif string(letter) == "G"
            complement = string(complement,"C")
        else
            print("Illegal char, exit")
            close(infile)
            exit()
        end 
    end
    return complement
end

function get_reverse(sequence::String)
    reverse_sequence = ""
    for i in Base.reverse(1:length(sequence))
        reverse_sequence = string(reverse_sequence,sequence[i])
    end
    return reverse_sequence
end

output = open("output_8_1.txt", "w")

open("/Users/laurasansc/github/learning_julia/files/dna7.fsa") do file
    sequence = ""
    ID = ""
    counter = 0
    for line in eachline(file)
        if occursin(">", line)
            
            complement =  transcribe(sequence)    
            reverse_sequence = get_reverse(complement)
            
            write(output, "\n",ID, " ComplementStrand\n")
            for char in reverse_sequence
                write(output, char)
                counter += 1
                if counter == 60
                    write(output, "\n")
                    counter = 0
                end
                
            end
            counter = 0
            ID = line
        else
            sequence = string(sequence,line)
        end
    end
 
end

close(output)
```

**2. Improve on 2.10 by making a function that calculates the factorial. Add some input control (make sure you get positive integers, when you ask for a number).**

```julia
println("Enter number 1:")
num = parse(Int, readline())


function factorial(number::Int64)
    if number < 0
        println("The number entered is negative. ")
    else
        fact = 1

        while number > 0
            fact = fact * number
            number -= 1
        end
        print(fact)
    end
end

factorial(num)
```

**4. Calculate the the standard deviation (1.8355) of the numbers in ex1.dat. The formula leads to directly to a two-pass algorithm (looking at all the numbers twice), where you will have store all the numbers in memory in order to calculate SD. The inspired programmer will find a one-pass algorithm, where you calulate SD just by looking at a number once, thereby not using significant memory. The genius will explain why there is a difference between the two results.**

```julia
function get_sd(numbers::Array)
    n = length(numbers)
    
    average = sum(numbers)/n
    x = 0
    for number in numbers
        x += (number-average)^2
    end
    standard_deviation = sqrt(x/n)
    println(round(standard_deviation, digits =4))
end


open("/Users/laurasansc/github/learning_julia/files/ex1.dat") do file
    numbers = []
    for ln in eachline(file)
        ln = split(ln, "\t")
        for item in ln
            push!(numbers, parse(Float64, item))
        end
    end
    get_sd(numbers)
end
```

**5. Building upon exercise 8.4, which counted the number of occurrences of accession numbers in ex5.acc, now make the program print the list ordered by the occurrence. That means the accession numbers with most duplicates should be first, and accession numbers which only occurs once should be last in the list.**

```julia
acc_dict = Dict()
# make a dictionary that counts the duplicates
open("/Users/laurasansc/github/learning_julia/files/ex5.acc") do file
    for line in eachline(file)
        if haskey(acc_dict, line) == false
            acc_dict[line] = 1
        else
            acc_dict[line] += 1
        end
    end
end


occurrences = []
for (key, value) in acc_dict
    line = string(key, "\t", value)
    line = split(line) 
    push!(occurrences, line)
    #write(outfile, line)
end

occurrences = sort(occurrences, by = x -> x[2])
for entry in occurrences
    println(entry[1], "\t",entry[2])
end
```
